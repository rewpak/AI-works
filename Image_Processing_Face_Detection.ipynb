# Problem Descriptions
In this task, we'll use the Haar cascade face detector, a fast method introduced by Viola and Jones in 2001. Although newer technologies offer more accuracy, this detector is known for its quick face detection capabilities.

**Pattern Learning:** During its training phase, the Haar Cascade Classifier learns rectangular patterns to detect contrast variations. These are vital for recognizing different areas of a face.

**Cascade Structure:** The classifier consists of a series of detectors arranged in a cascade. It analyzes the image in small sections called sub-windows to find potential faces.

**Initial Rejection of Candidates:** As it scans the image, each small section or sub-window is assessed to determine if it might contain a part of a face. The first few detectors in the cascade are designed to quickly reject sub-windows that are unlikely to contain a face.

**Feature Detection:** In each sub-window that passes the initial stages, the classifier examines contrast differences to identify darker regions (like eyes and mouth) and lighter regions (like the forehead).

**Selective Processing of Candidates:** Only the most promising sub-windows, which show potential face-like features, are allowed to pass through the subsequent, more complex detectors in the cascade.

**Localization and Visualization:** When a sub-window successfully passes all the detection stages, it's marked as a face. The classifier then draws a bounding box around this area, localizing and visualizing the detected face within the image.

```
!pip install opencv-python
import cv2
from google.colab.patches import cv2_imshow
```

```
# Load the face detection model from the model file
!wget 'https://github.com/yongminli/data/raw/main/haarcascade_frontalface_default.xml'
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```
# Load the model file and image files to Google Colab
from google.colab import files
file = files.upload()
```

```
# Read an input image
img = cv2.imread("2.png")

# Convert it into a grayscale image
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Detect faces
# faces = face_cascade.detectMultiScale(image=gray, scaleFactor=1.2, minNeighbors=5, minSize=(50,50))
faces = face_cascade.detectMultiScale(image=gray, minNeighbors=5)

# Display the detecition results
for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 2)
cv2_imshow(img)
```
